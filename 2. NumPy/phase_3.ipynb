### np.swapaxes

numpy.swapaxes(a, axis1, axis2) efficiently interchanges two specified axes of a NumPy array, returning a view rather than a copy for high performance.

https://numpy.org/devdocs/reference/generated/numpy.swapaxes.html

```python
# Example: Swapping axes for a 2D array (rows and columns)
arr = np.array([[1, 2, 3], [4, 5, 6]])

# Shape (2, 3) -> Swapping axis 0 and 1 -> Shape (3, 2)
swapped = np.swapaxes(arr, 0, 1)

print(arr, "\n")
print(swapped)
```

### numpy.count_nonzero

The `numpy.count_nonzero()` function counts the number of non-zero elements in a given NumPy array. This function is useful for quickly assessing the density of data or counting valid entries in datasets.

https://numpy.org/doc/2.3/reference/generated/numpy.count_nonzero.html


```python
# Create a sample array
x = np.array([[0, 10, 20],
              [20, 30, 40],
              [0, 0, 0]])

# Count total non-zero elements in the entire array
total_nonzero = np.count_nonzero(x)
print(f"Total non-zero elements: {total_nonzero}") # Output: 5

# Count non-zero elements along axis 0 (columns)
nonzero_along_axis0 = np.count_nonzero(x, axis=0)
print(f"Non-zero elements along axis 0 (col): {nonzero_along_axis0}") # Output: [1 2 2]

# Count non-zero elements along axis 1 (rows)
nonzero_along_axis1 = np.count_nonzero(x, axis=1)
print(f"Non-zero elements along axis 1 (row): {nonzero_along_axis1}") # Output: [2 3 0]
```

### np.tile

**np.tile()** is a NumPy function that constructs a new array by repeating an input array `(A)` a specified number of times `(reps)`, effectively tiling it. It expands dimensions based on reps (e.g., (2, 3) repeats 2 times vertically and 3 times horizontally) and is essential for data broadcasting in machine learning.

https://numpy.org/doc/2.3/reference/generated/numpy.tile.html

```python
a = np.array([0, 1, 2])
np.tile(a, 2)
```

```python
np.tile(a, (2, 2))
```

```python
np.tile(a, (2, 1, 2))
```

```python
b = np.array([[1, 2], [3, 4]])
np.tile(b, 2)
```

```python
np.tile(b, (2, 1))
```

### np.allclose

**numpy.allclose()**

This function is designed to handle the inherent precision limitations of floating-point arithmetic. It returns True if two arrays are element-wise equal within a combination of relative and absolute tolerances.

https://numpy.org/doc/2.1/reference/generated/numpy.allclose.html

```python
np.allclose([1e10,1e-7], [1.00001e10,1e-8])
```

```python
np.allclose([1e10,1e-8], [1.00001e10,1e-9])
```

```python
np.allclose([1e10,1e-8], [1.0001e10,1e-9])
```

```python
np.allclose([1.0, np.nan], [1.0, np.nan])
```

```python
np.allclose([1.0, np.nan], [1.0, np.nan], equal_nan=True)
```
